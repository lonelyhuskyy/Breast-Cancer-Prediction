```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
cancer_data = datasets.load_breast_cancer()
cancer_data
```

```python
print("Features : ",cancer_data.feature_names)
```

```python
df = pd.DataFrame(data=cancer_data.data,columns=cancer_data.feature_names)
df['Target'] = cancer_data.target
df
```

```python
df.describe()
```

```python
df.info()
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
fig,ax = plt.subplots(nrows=10,ncols=3,figsize=(30,90))

for i,column in enumerate(df.columns[:-1]):
    ax_i = ax[i // 3, i % 3]
    ax_i.scatter(df[column],df['Target'])
    ax_i.set_xlabel(column)
    ax_i.set_ylabel('Target')
    ax_i.set_title(f'{column} vs Target')
plt.tight_layout()
plt.show()
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X = sc_x.fit_transform(X)
```

# Checking Correlation

```python
corr = df.corr()
corr
```

```python
import seaborn as sns
fig,ax = plt.subplots(figsize=(20,20))
sns.heatmap(data=corr,annot=True, ax=ax)
plt.show()
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=7)
pca.fit(X)
```

```python
pca.explained_variance_
```

```python
sum(pca.explained_variance_)
```

Great, we've preserved only 27.3509% of covariance after applying PCA

```python
X_red = pca.transform(X)
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test_cv, y_train, y_test_cv = train_test_split(X_red,y,test_size=0.3,random_state=35)
X_test, X_cv, y_test, y_cv = train_test_split(X_test_cv,y_test_cv,test_size=0.5,random_state=20)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(penalty='l2')
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test_cv)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
cm = confusion_matrix(y_pred,y_test_cv)
acc = accuracy_score(y_test_cv,y_pred)
f1 = f1_score(y_test_cv,y_pred)
```

```python
print(f"Confusion Matrix : \n",cm)
print(f"Accuracy : ",acc)
print(f"F1 score : ",f1)
```

# K-Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train,y_train)
```

# XGBOOST Classifier 

```python
param_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 200, 300],
    'gamma': [0, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV

classifier = xgb.XGBClassifier()


grid_search = GridSearchCV(estimator=classifier, param_grid=param_grid, scoring='roc_auc', cv=5)
grid_search.fit(X_train, y_train)

print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_cv)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
cm = confusion_matrix(y_pred,y_test_cv)
acc = accuracy_score(y_test_cv,y_pred)
f1 = f1_score(y_test_cv,y_pred)
```

```python
print(f"Confusion Matrix : \n",cm)
print(f"Accuracy : ",acc)
print(f"F1 score : ",f1)
```

# ANN Binary Classification

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(units=16,activation='relu'),
    tf.keras.layers.Dense(units=8,activation='relu'),
    tf.keras.layers.Dense(units=1,activation='sigmoid')
])
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_cv,y_cv),epochs=50,batch_size=32,verbose=1)
```

```python
y_pred_prob = model.predict(X_test)
```

```python
threshold = 0.5
y_pred = (y_pred_prob >= threshold).astype(int)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
cm = confusion_matrix(y_pred,y_test)
acc = accuracy_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
```

```python
print(f"Confusion Matrix : \n",cm)
print(f"Accuracy : ",acc)
print(f"F1 score : ",f1)
```
